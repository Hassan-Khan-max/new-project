<h2> Now Scraping the 50 Pages </h2>

```python
Book_list=[]
map_rating={'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
for i in range(1,51):
    url = f'https://books.toscrape.com/catalogue/page-{i}.html'
    web = requests.get(url)
    soup = BeautifulSoup(web.text, "html.parser")
    books = soup.find_all('article', class_='product_pod')
    for book in books:
        Title = book.h3.a['title']
        book_rating = map_rating[book.p['class'][1]]
        price = book.find('p',class_='price_color').text
        Availibilty  = book.find('p', class_='instock').text.strip()
        Book_list.append([Title, book_rating, price, Availibilty])

        
```

```python
len(Book_list)
```

```python
Books_df = pd.DataFrame(Book_list)
```
